We will study the data set of 3d acceleration for three walkers, which can be found in https://github.com/GUDHI/TDA-tutorial

It has been obtained from the walk of 3 persons A, B and C. 

The walk has been recorded using the accelerometer sensor of a smartphone in their pocket, giving rise to 3 time series, each of them representing the 3 coordinates of the acceleration of the corresponding walker in a coordinate system attached to the sensor. 

Each serie has been splitted in a list of 100 times series made of 200 consecutive coordinate points.

```python
f = open("data_acc","rb")
data = pickle.load(f)
f.close()
```

```python
data_A_sample = data[0][0]
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
ax.scatter(data_A_sample[:, 0], data_A_sample[:, 1], data_A_sample[:, 2]);
```

## Rips complex and persistence diagram

For each time series of each walker we can produce a barcode and a persistence diagram from the Rips filtration of the points. We show this for the first trajectory of the first walker A.

We begin by building the Rips complex, setting the radius to 0.2 maximum and 3 as the maximum dimension of the simplexes. 

Then, we can obtain a list where each element is a tuple (dim,(b,d)): dim is the dimension of the homology group, b is the birth parameter and d is the death parameter. 

These can be represented as intervals $(b,d)$ in the case of the barcode, or as points in the plane $(b,d)\in\R^2$ in the case of the persistence diagram.

```python
#1-skeleton of the Rips Complex, pairing pòints at distance less or equal than 0.2
skeletonA0 = gd.RipsComplex(points = data_A_sample, max_edge_length = 0.2);

#Compute Rips Complex
Rips_simplex_treeA0= skeletonA0.create_simplex_tree(max_dimension = 3);

#Obtain barcode
BarCodes_RipsA0 = Rips_simplex_treeA0.persistence();
for i in range(10):
    print(BarCodes_RipsA0[i])
```

```python
#Plot barcodes
gd.plot_persistence_barcode(BarCodes_RipsA0, legend=True)
```

```python
#Plot persistence diagram
gd.plot_persistence_diagram(BarCodes_RipsA0,legend=True);
```

## Vectorizations

This persistence diagrams can be transformed, via a vectorization, to an element of a vector space. We show this for the persistence diagram that is obtained from the first time series of walker A. We focus only on the first homology group. In particular, we produce the persistence landscape, silhouette and Betti curve.

```python
#Compute persistence barcode corresponding only to the
#first homology group
Rips_dim1A0 = Rips_simplex_treeA0.persistence_intervals_in_dimension(1);

#Remove essential part
Ess = gd.representations.preprocessing.DiagramSelector(use = True, point_type='finite')
Rips_dim1A0_NoEss = Ess.fit_transform([Rips_dim1A0]);

#Obtain first 5 landscapes
LS = gd.representations.Landscape(resolution=1000)
LandscapeA0 = LS.fit_transform([Rips_dim1A0_NoEss][0])

#plot first 3 landscapes
plot(LandscapeA0[0][:1000])
plot(LandscapeA0[0][1000:2000])
plot(LandscapeA0[0][2000:3000])
#xlabel('foo')
#ylabel('bar')
legend(['1st', '2nd','3rd'])
title('Persistence Landscape')

print(LandscapeA0.shape)
```

```python
#Obtain persistence silhouette weighted by distance to the diagonal
SH = gd.representations.Silhouette(resolution=1000, weight=lambda x: np.power(x[1]-x[0],1))
ShilouetteA0 = SH.fit_transform([Rips_dim1A0_NoEss][0])

plt.plot(ShilouetteA0[0])
title('Persistence Silhouette')

print(ShilouetteA0.shape)
```

```python
#Obtain Betti curve
BC = gd.representations.vector_methods.BettiCurve(resolution=1000)
BettiCurveA0 = BC.fit_transform([Rips_dim1A0_NoEss][0])

plot(BettiCurveA0[0])
title('Betti Curve')

print(BettiCurveA0.shape)
```

## Principal component analysis

Thanks to vectorizations we can transform persistence diagrams into elements of a vector space, and then apply linear methods as principal component analysis (PCA). PCA reduces the dimension of the data provided as input by projecting each point onto only the first few principal components (orthogonal vectors that best fit the data). This let us visualize high dimensional data sets while preserving as much of the data's variation as possible.

In the following we are going to perform PCA on some vectorizations of the data set of walkers. We focus on one dimensional homology groups. We will study if, with the information provided, we are able to differentiate the three walkers.

For each walker we have 100 time series. Each of them provides a persistence diagram, which in turn is transformed into a vector of lenght $300\ll m$. Therefore, since there are three walkers, we obtain a matrix of dimension $300\times m$ that gather all this information: each line is the vector associated to the corresponding persistence diagram of that time series.

We proceed as follows:

- compute PCA to reduce the dimension of the matrix to $300\times 300$, 
- from here we just take the two first coordinates of each line and represent them in the plane. 

In this way, the topological information of each trajectory is reduced to a point in $\mathbb{R}^2$.

Blue is walker A, red is B and green is C.

```python
from sklearn.decomposition import PCA
pca = PCA()
```

We start with **Betti curves**.

```python
#Compute Betti curves for each time series
Betti = np.zeros((1,1000))
for i in range(3):
    for j in range(100):
        skeleton = gd.RipsComplex(points = data[i][j], max_edge_length = 0.2);
        Rips_simplex_tree= skeleton.create_simplex_tree(max_dimension = 3);
        Rips_simplex_tree.persistence();
        Rips_dim1 = Rips_simplex_tree.persistence_intervals_in_dimension(1);
        Rips_dim1_NoEss = Ess.fit_transform([Rips_dim1]);
        BettiCurve = BC.fit_transform([Rips_dim1_NoEss][0]);
        Betti = np.concatenate((Betti, BettiCurve));
        
Betti = np.delete(Betti, 0, 0)
```

```python
np.shape(Betti)
```

```python
#Compute PCA
BettiPCA = pca.fit(Betti).transform(Betti)
print(BettiPCA.shape)
```

```python
# 2D Projection of the data
plt.figure(figsize=(7,7))
for i, j in zip(BettiPCA[:100,0], BettiPCA[:100,1]):
    plt.text(i, j, "o", color="blue")
for i, j in zip(BettiPCA[100:200,0], BettiPCA[100:200,1]):
    plt.text(i, j, "o", color="red")
for i, j in zip(BettiPCA[200:300,0], BettiPCA[200:300,1]):
    plt.text(i, j, "o", color="green")    
 
plt.axis((-150,150,-150,150))  
plt.show()

#Blue is walker A, red is B and green is C.
```

Then, we proceed analogously for **landscapes**.

```python
#Compute the first 5 landscapes for each time series.

#We transform into a zero vector those persistence diagrams
#that only have essential points.

LandS = np.ones((1,5000))
for i in range(3):
    for j in range(100):
        skeleton = gd.RipsComplex(points = data[i][j], max_edge_length = 0.2);
        Rips_simplex_tree= skeleton.create_simplex_tree(max_dimension = 3);
        Rips_simplex_tree.persistence();
        Rips_dim1 = Rips_simplex_tree.persistence_intervals_in_dimension(1);
        Rips_dim1_NoEss = Ess.fit_transform([Rips_dim1]);
        if Rips_dim1_NoEss[0].size == 0:
            Landscape = np.zeros((1,5000))
        else:
            Landscape = LS.fit_transform([Rips_dim1_NoEss][0])
        
        LandS = np.concatenate((LandS, Landscape));
        
LandS = np.delete(LandS, 0, 0)

np.shape(LandS)
```

```python
#Compute PCA
LandSPCA = pca.fit(LandS).transform(LandS)

print(LandSPCA.shape)
```

```python
# 2D Projection of the data
plt.figure(figsize=(7,7))
for i, j in zip(LandSPCA[:100,0], LandSPCA[:100,1]):
    plt.text(i, j, "o", color="blue")
for i, j in zip(LandSPCA[100:200,0], LandSPCA[100:200,1]):
    plt.text(i, j, "o", color="red")
for i, j in zip(LandSPCA[200:300,0], LandSPCA[200:300,1]):
    plt.text(i, j, "o", color="green")    
 
plt.axis((-1,1,-1,1))  
plt.show()

#Blue is walker A, red is B and green is C.
```

We repeat the previous computation but just with the 1-landscapes of each diagram.

```python
#Compute landscapes but choosing only the first one for PCA
LandS1 = np.zeros((1,1000))
for i in range(3):
    for j in range(100):
        skeleton = gd.RipsComplex(points = data[i][j], max_edge_length = 0.2);
        Rips_simplex_tree= skeleton.create_simplex_tree(max_dimension = 3);
        Rips_simplex_tree.persistence();
        Rips_dim1 = Rips_simplex_tree.persistence_intervals_in_dimension(1);
        Rips_dim1_NoEss = Ess.fit_transform([Rips_dim1]);
        if Rips_dim1_NoEss[0].size == 0:
            Landscape = np.zeros((1,5000));
        else:
            Landscape = LS.fit_transform([Rips_dim1_NoEss][0]);
        
        LandS1 = np.concatenate((LandS1, [Landscape[0][:1000]]));
        
LandS1 = np.delete(LandS1, 0, 0)
np.shape(LandS1)
```

```python
#Compute PCA
LandS1PCA = pca.fit(LandS1).transform(LandS1)

print(LandS1PCA.shape)
```

```python
# 2D Projection of the data
plt.figure(figsize=(7,7))
for i, j in zip(LandS1PCA[:100,0], LandS1PCA[:100,1]):
    plt.text(i, j, "o", color="blue")
for i, j in zip(LandS1PCA[100:200,0], LandS1PCA[100:200,1]):
    plt.text(i, j, "o", color="red")
for i, j in zip(LandS1PCA[200:300,0], LandS1PCA[200:300,1]):
    plt.text(i, j, "o", color="green")    
 
plt.axis((-.8,.8,-.8,.8))  
plt.show()

#Blue is walker A, red is B and green is C.
```

From the computations above we can deduce that the Betti curves differentiate the three walkers better than the landscapes, and choosing just the first landscape does not improve the classification. Moreover, it can be seen how the time series of walker C (green) stands out above the ones of the other two walkers, which are more intermingled.

Let us study a little more what are we obtaining with PCA.

First, we focus on the **Betti curves**.

```python
#Compute PCA for Betti curves
#If 0<n_components<1, it means the amount of variance 
#that needs to be explained is greater than n_components%

pca_Betti = PCA(n_components=.8)
pcaB = pca_Betti.fit_transform(Betti)

shape(pcaB)
```

```python
#Percentage of variance explained by each of the components
print(pca_Betti.explained_variance_ratio_)
```

```python
#Clustering of the data after PCA using KMeans
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3)

#Labels of each point
pcaBetti_labels = kmeans.fit(pcaB).labels_
[pcaBetti_labels[:100],pcaBetti_labels[100:200], pcaBetti_labels[200:300]]

#We see how walker C is extremely well classified. 
```

We move onto **landscapes**.

```python
#Compute PCA for landscapes

pca_LandS = PCA(n_components=.8)
pcaL = pca_LandS.fit_transform(LandS)

shape(pcaL)
```

```python
#Percentage of variance explained by each of the components
print(pca_LandS.explained_variance_ratio_)
```

```python
#Clustering of data after PCA using KMeans
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3)

#Labels of each point
pcaLandS_labels = kmeans.fit(pcaL).labels_
[pcaLandS_labels[:100],pcaLandS_labels[100:200], pcaLandS_labels[200:300]]

#Again, walker 3 is the best differentiated, although not as good as with Betti curves
#It seems that landscapes differentiate worst than Betti curves
```
